```python
import json
import re

with open('batch1_20docs_export.json', "r", encoding="utf-8") as f:
    tasks = json.load(f)

def get_tokens(text):
    return list(re.finditer(r"\S+", text))

def make_binary_labels(text, results):
    tokens = get_tokens(text)
    labels = [0] * len(tokens)

    for item in results:
        if item["type"] == "labels":
            start = item["value"]["start"]
            end = item["value"]["end"]

            for i, tok in enumerate(tokens):
                if tok.start() < end and tok.end() > start:
                    labels[i] = 1

    return labels

annotator_a = []
annotator_b = []

for task in tasks:
    text = task["data"]["text"]
    annotations = task["annotations"]

    if len(annotations) != 2:
        continue

    labels_a = make_binary_labels(text, annotations[0]["result"])
    labels_b = make_binary_labels(text, annotations[1]["result"])

    annotator_a.extend(labels_a)
    annotator_b.extend(labels_b)

print(f"Annotator A: {annotator_a}")
print(f"Annotator B: {annotator_b}")
```

```python
##code for calculating IAA for the initial batch of annotations
```

```python
from sklearn.metrics import cohen_kappa_score, f1_score


kappa=cohen_kappa_score(annotator_a, annotator_b)

f1=f1_score(annotator_a,annotator_b)

print(f"Kappa: {kappa:3f}, F1: {f1:3f}")
```
